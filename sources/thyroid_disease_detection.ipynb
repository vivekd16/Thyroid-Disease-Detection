```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dataset/raw/thyroidDF.csv')
df
```

```python
df.info()
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
ohe=LabelEncoder()
categorical_columns=['sex', 'on_thyroxine', 'query_on_thyroxine',
       'on_antithyroid_meds', 'sick', 'pregnant', 'thyroid_surgery',
       'I131_treatment', 'query_hypothyroid', 'query_hyperthyroid', 'lithium',
       'goitre', 'tumor', 'hypopituitary', 'psych', 'TSH_measured',
       'T3_measured', 'TT4_measured', 'T4U_measured',
       'FTI_measured', 'TBG_measured', 'referral_source',]
for i in categorical_columns:
       df[i]=ohe.fit_transform(df[i])
df
```

```python
hyperthyroid_conditions = ['A', 'B', 'C', 'D', 'O', 'P', 'Q', 'S', 'T']
hypothyroid_conditions = ['E', 'F', 'G', 'H']
normal_conditions = ['-']

def categorize_target(value):
    # Split the diagnosis into parts for compound cases
    diagnoses = value.split('|')
    # Check for hyperthyroid conditions
    for diagnosis in diagnoses:
        if diagnosis in hyperthyroid_conditions:
            return 1 #for hyperthyroid value 
    # Check for hypothyroid conditions
    for diagnosis in diagnoses:
        if diagnosis in hypothyroid_conditions:
            return 2 #for hypothyroid value
            
    for diagnosis in diagnoses:
        if diagnosis in normal_conditions:
            return 0 #for the negative value

#Applying 'categorize_target' function on 'target' column to categorize the values into 3 classes
df['target'] = df['target'].apply(categorize_target)
```

```python
import seaborn as sns
hmdf=df.corr()
plt.figure(figsize=(100,50))
sns.heatmap(hmdf)
```

```python
df["target"].unique()
```

```python
print(df['target'].isnull().sum())
```

```python
df = df[df['target'].notna()]
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_temp,y_train,y_temp=train_test_split(df.drop("target",axis=1),df["target"],random_state=42,test_size=0.4)
X_test,X_val,y_test,y_val=train_test_split(X_temp,y_temp,random_state=42,test_size=0.5)
```

```python
print(X_train.shape)
print(X_test.shape)
X_val.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_val)
print(accuracy_score(y_pred,y_val))
print(classification_report(y_val, y_pred))
print(confusion_matrix(y_val, y_pred))
```

```python
y_pred=model.predict(X_test)
print(accuracy_score(y_pred,y_test))
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
importances = model.feature_importances_
feature_names = X_train.columns
feature_importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
print(feature_importance_df)
```

```python
new_df = df.drop("hypopituitary",axis=1).drop("referral_source",axis=1).drop("patient_id",axis=1)
# new_df.to_csv('dataset\\ProcessedthyroidDF.csv',index=False)

from sklearn.model_selection import train_test_split

X_train,X_temp,y_train,y_temp=train_test_split(new_df.drop("target",axis=1),new_df["target"],random_state=42,test_size=0.4)
X_test,X_val,y_test,y_val=train_test_split(X_temp,y_temp,random_state=42,test_size=0.5)

print(X_train.shape)
print(X_test.shape)
print(X_val.shape)

```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report, confusion_matrix
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
}
grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=3)
grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
```

```python
y_pred=grid_search.predict(X_test)
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Predict probabilities for all classes
y_pred_proba = grid_search.predict_proba(X_test)

# Compute multi-class ROC-AUC
roc_auc = roc_auc_score(y_test, y_pred_proba, multi_class='ovr')  # Use 'ovo' or 'ovr'
print(f"Multi-Class ROC-AUC Score: {roc_auc}")

classes = np.unique(y_test)
y_test_bin = label_binarize(y_test, classes=classes)

plt.figure(figsize=(10, 8))

# Compute ROC curve for each class
for i, class_label in enumerate(classes):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_pred_proba[:, i])
    plt.plot(fpr, tpr, label=f"Class {class_label} ROC Curve (AUC = {roc_auc:.2f})")

plt.plot([0, 1], [0, 1], 'k--', label="Random Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve for Multi-Class Classification")
plt.legend()
plt.show()
```
